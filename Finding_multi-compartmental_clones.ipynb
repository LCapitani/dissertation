```python
import pandas as pd 
```

```python
ad = pd.read_csv("all_data.tab", sep = "\t")
```

```python
ad['PATIENT'].unique()
```

```python
multi_compartment_clones = pd.DataFrame(columns=['patient','clone'])

for patient in ad.loc[ad['LOCATION'] != 'Healthy_Periphery', 'PATIENT'].unique():
    patient_data = ad.loc[ad['PATIENT'] == patient]
    for clone in patient_data['CLONE'].unique():
        clone_data = patient_data.loc[patient_data['CLONE'] == clone]
        locations = clone_data['LOCATION'].unique()
        if len(locations) > 1:
            tempdf = pd.DataFrame(columns=['patient','clone'])
            tempdf.at[0, 'patient'] = patient
            tempdf.at[0, 'clone'] = clone
            multi_compartment_clones = multi_compartment_clones.append(tempdf)
        else:
            continue
```

```python
multi_compartment_clones['clone'].iloc[1]
```

```python
 multi_compartment_clones['seqnum'] = ''

for i in range(len(multi_compartment_clones)):
    clone = multi_compartment_clones['clone'].iloc[i]
    patient = multi_compartment_clones['patient'].iloc[i]
    patient_data = ad.loc[ad.PATIENT == patient]
    clone_data = patient_data.loc[patient_data.CLONE == clone]
    length = len(clone_data)
    multi_compartment_clones['seqnum'][i:i+1] = length
```
